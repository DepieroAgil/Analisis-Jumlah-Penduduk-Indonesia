```python
# Import pustaka yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat dataset sederhana dengan data jumlah penduduk kota besar di Indonesia
data = {
    'Kota': ['Jakarta', 'Surabaya', 'Medan', 'Bandung', 'Bekasi', 'Depok', 'Tangerang', 'Semarang', 'Palembang', 'Makassar'],
    'Jumlah_Penduduk': [10850634, 2875420, 2308144, 2394870, 2748219, 2098031, 1762984, 1688833, 1600000, 1456188]  # Data jumlah penduduk (estimasi)
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Menampilkan informasi dasar dari dataset
print("Informasi dataset:")
print(df.info())
print("\nStatistik deskriptif:")
print(df.describe())

# Menampilkan DataFrame
print("\nData kota dengan jumlah penduduk terbanyak di Indonesia:")
print(df)

# Visualisasi data dengan barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Kota', y='Jumlah_Penduduk', data=df, palette='viridis')
plt.title('Jumlah Penduduk Kota-Kota Besar di Indonesia')
plt.xlabel('Kota')
plt.ylabel('Jumlah Penduduk')
plt.xticks(rotation=45)
plt.show()

# Visualisasi distribusi data dengan histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Jumlah_Penduduk'], bins=10, kde=True, color='skyblue')
plt.title('Distribusi Jumlah Penduduk')
plt.xlabel('Jumlah Penduduk')
plt.ylabel('Frekuensi')
plt.show()
```
